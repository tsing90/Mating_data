## check if .png is a real RGBA image

```python
import cv2
import os
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np
```

```python
ait_path = '/home/kevin/git/Semantic_Human_Matting/data/ait_list.txt'
bg_path = '/home/kevin/git/Semantic_Human_Matting/data/bg_list.txt'
dim_path = '/home/kevin/git/Semantic_Human_Matting/data/DIM_list.txt'
```

```python
badList = []
with open(ait_path,'r') as f:
    paths = f.readlines()
#paths = paths[::-1]
for path in paths:
    img = cv2.imread(path.strip(), -1)
    alpha = img[:,:,3]
    if np.max(alpha) <= 250:
        print(path.strip())
        badList.append(path)
```

```python
for path in badList:
    img = cv2.imread(path.strip(),-1)
    alpha = img[:,:,3]

    if np.max(alpha)<=250:
        alpha[alpha == np.max(alpha)] = 255
    
    img[:,:,3] = alpha
    
    cv2.imwrite(path.strip(), img)
```
